```python
import numpy as np 
import pandas as pd
from pandas import Series,DataFrame
```

```python
df1 = DataFrame({'name':['A','B','C','D','E','F'],
                 'sex':['男','女','女','女','男','男'],
                 'id':[1,2,3,4,5,6]},index=range(101,107))
df1
```

```python
df2 = DataFrame({'age':[22,25,27,21,18,29],'salary':[12000,15000,20000,30000,10000,8000],'id':[1,2,3,4,5,7]},index=range(101,107))
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1.append(df2)
```

```python
df1.merge(df2,how='right')
```

```python
df = DataFrame(np.random.randint(0,150,size=(6,3)),index=list('ABCDEF'),columns=['python','en','math'])
df
```

```python
s = df.mean().round(1)
s
```

```python
df3 = DataFrame(s)
df3.columns = ['sorce']
df3
```

```python
df5 = df3.T
df5
```

```python
df6 = df.append(df5)
df6
```

```python
df4 = df3.unstack()
df4
# DataFrame(df4).stack()
```

```python
df7 = df6.mean(axis=1).round(1)
df7 = DataFrame(df7,columns=['mean'])
df7
```

```python
#left_index=True, right_index=True:使用左/右边的索引列作为链接键
df6.merge(df7,how='outer',left_index=True, right_index=True)
```
