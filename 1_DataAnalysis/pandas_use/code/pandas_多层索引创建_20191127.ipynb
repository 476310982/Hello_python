```python
import numpy as np
import pandas as pd
#数据分析BI-->人工智能AI
#数据分析和数据挖掘是一个意思
#工具和软件：Excel/SPSS/SAS/Matlab
from pandas import Series,DataFrame
```

```python
#单层索引
s1 = Series(data=np.random.randint(0,150,size=6),index=list('abcdef'),dtype=np.int16,name='single index')
s1
```

```python
#多层索引
s2 = Series(data=np.random.randint(0,150,size=6),index=pd.MultiIndex.from_product([['zs','ls','ww'],['期中','期末']]),dtype=np.int16,name='double index')
s2                                                                              
```

```python
df = DataFrame(data=np.random.randint(0,150,size=(10,4)),index=range(100,110),columns=['python','en','math','music'])
df
```

```python
df1 = DataFrame(data=np.random.randint(0,150,size=(12,4)),index=pd.MultiIndex.from_product([['zs','ls','ww'],['期中','期末'],['A','B']]),columns=['python','en','math','music'])
df1
```

```python
#按列查找
df1['python']['zs']['期中']['A']
```

```python
#按行（索引）查找
df1.loc['zs','期中','A']['en']
```

```python
df1.iloc[5]
```
