```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df = DataFrame(data=np.random.randint(0,150,size=(6,6)),index=list('ABCDEF'),columns=pd.MultiIndex.from_product([['python','en','math'],['期中','期末']]))
df
```

```python
df.mean().round(2)
```

```python
#level代表索引位置,注意axis
df.mean(axis=1,level=0)
```

```python
df.mean(axis=1,level=1).round(1)
```

```python
#行和列的多层索引，进行转换
#stack:将指定的级别从列堆叠到索引
#这里的level
df2 = df.stack(level=0)
df2
```

```python
df2.unstack(level=0)
```

```python
df2.unstack(level=-1)
```
