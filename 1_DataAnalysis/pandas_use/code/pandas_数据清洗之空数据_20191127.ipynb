```python
import numpy as np
import pandas
from pandas import Series,DataFrame
```

```python
#判断是否有空数据 isnull :存在空数据返回True，否则为False
```

```python
df = DataFrame(data=np.random.randint(0,150,size=(10,5)),index=range(100,110),columns=['python','en','math','chinese','music'])
df.isnull()
```

```python
df.notnull().all()
```

```python
'''
python的内置函数之二
元素除了是 0、空、FALSE 外都算 TRUE。
xxx.any()  any() 函数用于判断给定的可迭代参数 iterable 是否全部为 False，则返回 Fals
           如果有一个为 True，则返回 True。
xxx.all()  函数用于判断给定的可迭代参数 iterable 中的所有元素是否都为 TRUE
           如果是返回 True，否则返回 False。
'''
```

```python
#获得随机行数（索引）
for i in range(10):
    index = np.random.randint(100,110,size=1)[0]
    col = np.random.choice(df.columns)
#     print("index:%d col:%s value:%d"%(index,col,df.loc[index,col]))
#     df.loc[index,col] = None
#     df.loc[index,col] = np.NAN
    df.loc[index,col] = np.NaN
    #三个代码效果一样
df
```

```python
df.isnull().any()#存在空值
df.notnull().all()#存在空值
```

```python
'''
df2.fillna(
    value=None,
    method=None,
    axis=None,
    inplace=False,
    limit=None,
    downcast=None,
    **kwargs,)
    
inplace参数的取值：True、False
    True：直接修改原对象
    False：创建一个副本，修改副本，原对象不变（缺省默认）

method参数的取值 ： {‘pad’, ‘ffill’,‘backfill’, ‘bfill’, None}, default None
    pad/ffill：用前一个非缺失值去填充该缺失值
    backfill/bfill：用下一个非缺失值填充该缺失值
    None：指定一个值去替换缺失值（缺省默认这种方式）
'''
```
