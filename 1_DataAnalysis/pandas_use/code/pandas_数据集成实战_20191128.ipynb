```python
import numpy as np 
import pandas as pd
from pandas import Series,DataFrame
```

```python
#分析美国人口数据
#pandas读取csv数据
#pandas中有很多读取数据的方法

#读取美国大洲数据
areas = pd.read_csv('../source/state-areas.csv')
type(areas)#pandas.core.frame.DataFrame
areas
```

```python
#读取美国大洲缩写数据
abbrevs = pd.read_csv('../source/state-abbrevs.csv')
type(abbrevs)#pandas.core.frame.DataFrame
abbrevs
```

```python
display(areas.shape,abbrevs.shape)
```

```python
#读取美国各州人口总量数据
population = pd.read_csv('../source/state-population.csv')
type(population)#pandas.core.frame.DataFrame
population
```

```python
display(population.head(),abbrevs.head())
```

```python
#以相同值作为级联键进行关联
#left_on='state/region',right_on='abbreviation' 声明左/右边关联的列索引键
pop = population.merge(abbrevs,how='outer',left_on='state/region',right_on='abbreviation')
pop
```

```python
pop.shape
```

```python
#查找含有空值的列
pop.isnull().any()
```

```python
#删除重复的列（表示意义相同）
pop.drop(labels='abbreviation',axis=1,inplace=True)
```

```python
pop.head()
```

```python
#定位人口为空的列
pop.isnull().any()
# type(cond)#pandas.core.series.Series
```

```python
#筛选出空值的index（True）
# type(cond)#pandas.core.series.Series
cond = pop['state'].isnull()
cond
```

```python
#筛选出空值index的'state/region'值并去重，得出缺失的大洲缩写
pop[cond]['state/region'].unique()
```

```python
#得出'PR'空值的index
cond = pop['state/region']=='PR'
cond.shape
```

```python
#pandas的高级检索：将为True的index对应值进行替换
pop['state'][cond] = 'Puerto Rico'
```

```python
#pandas的高级检索：同上
cond = pop['state/region']=='USA'
pop['state'][cond] = 'United State'
```

```python
#验证结果，'state'列不存在空值
pop.isnull().any()
```
