```python
#计算机视觉 opencv
# open computer vision
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#cv2模块下，都是使用python方法调用c++
#所以导包的时候，导入cv2就可以了，使用相应的方法了
import cv2
#cv2读取图片时，路径不能带中文
#cv2读取图片是，颜色通道为蓝绿红（正常情况时红绿蓝）
```

```python
rose = cv2.imread('../source/rose.jpg')
rose.shape
```

```python
plt.imshow(rose[:,:,::-1])
```

```python
#修改图片颜色
gray = cv2.cvtColor(rose,code = cv2.COLOR_BGR2GRAY)
#设图片大小
img = cv2.resize(gray,dsize = (200,100))

#cv2可以显示图片
cv2.imshow('rose',img)
#等待键盘输入“q”,否则无限等待
#cv2.waitKey(0) 单位毫秒 会返回 键盘输入的ASICLL码值
while True:
    if ord('q') == cv2.waitKey(100):
        break

#释放内存
cv2.destroyAllWindows()
```
